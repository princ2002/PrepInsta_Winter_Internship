# df (working with data frame)

### pd.melt() used to convert columns into row 

```python
df =    pd.melt(population,
        id_vars='country_code',
        value_vars = years,
        var_name = 'Year',
        value_name = 'population'
        
)
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.country_code.value_counts().count()
```

```python
country.head()
```

```python
df.head()
```

```python
df_merged = pd.merge(country,df,
         how = 'left',
         on= 'country_code')
df_merged.head()
```

```python
df_merged.shape
```

### Fertility data 


Reshape the fertility data 

```python
df_fertility = pd.melt(fertility,
                id_vars= 'country_code',
                value_vars= years,
                var_name = 'Year',
                value_name= 'fertility',
                )
df_fertility.head()
```

Merge the df_fertility with df_merged 

```python
df_merged = pd.merge(df_merged,df_fertility,
            how = 'left',
            on = ['country_code','Year']
            )
df_merged.head()
```

### Life Expectancy 

#### Reshape the life expectancy data 

```python
df_life_exp =   pd.melt(life_exp,
                id_vars='country_code',
                value_vars=years,
                var_name= 'Year',
                value_name= 'life_exp'
        )
df_life_exp.head()
```

```python
df_merged.head()
```

### Merge df_life_exp with df_merged

```python
df_merged = pd.merge(df_merged,df_life_exp,
            how = 'left',
            on = ['country_code','Year']
            )
df_merged.head()
```

```python
df_merged.shape
```
