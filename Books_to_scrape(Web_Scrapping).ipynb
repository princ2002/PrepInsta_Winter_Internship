# Books to scrape 

```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#url = 'https://books.toscrape.com/'
url = "https://books.toscrape.com/catalogue/page-1.html"
r = requests.get(url)
print(r.status_code)
```

```python
soup = BeautifulSoup(r.text,'lxml')
#print(soup)
```

```python
ol = soup.find("ol",class_="row")
articles = ol.find_all("article",class_="product_pod")
articles
```

```python
title_list = []

for article in articles:
    image = article.find("img")
    title = image.attrs["alt"]
    title_list.append(title)
 
for title in title_list:   
    print(title)

    
```

```python
title_list = []
Review = []
Price = []

for article in articles:
    image = article.find("img")
    title = image.attrs["alt"]
    title_list.append(title)
    #print(title)
    
    star = article.find("p")
    star = star['class'][1]
    Review.append(star)
    #print(star)
    
    price = article.find("p",class_="price_color").text
    price = price[2:]  ## Remove the first two unnecassary elements of the price 
    price = float(price) ## Convert the price into float 
    #print(price)
    Price.append(price)
    
 
#for title in title_list:   
    #print(title)
```

## DataFrame

```python
df = pd.DataFrame({"Title":title_list,"Price":Price,"Review":Review})
df.head()
```

### Extract all 20 Pages 

```python
Books = []   
 
for i in range(1,21):
    url = f"https://books.toscrape.com/catalogue/page-{i}.html"
    r = requests.get(url)
    #print(r.status_code)
    
    soup = BeautifulSoup(r.text,'lxml')
    ol = soup.find("ol",class_="row")
    articles = ol.find_all("article",class_="product_pod")

    
    #Books = [] Add it in outside the loop bcuz as the loop run Books print empty everytime 

    for article in articles:
        image = article.find("img")
        title = image.attrs["alt"]
        #title_list.append(title)
        #print(title)
        
        star = article.find("p")
        star = star['class'][1]
        #Review.append(star)
        #print(star)
        
        price = article.find("p",class_="price_color").text
        price = price[2:]  ## Remove the first two unnecassary elements of the price 
        price = float(price) ## Convert the price into float 
        #print(price)
        #Price.append(price)
        
        link = article.find("a").get("href")
        links = "https://books.toscrape.com/catalogue/" + link
        #print(links)
        
        Books.append([title,price,star,links])
        
print("Total no of books in",i,"pages are",len(Books))
```

```python
df = pd.DataFrame(Books,columns=["Title","Prices","Reviews","Links"])
df.head()
#df.shape
```
