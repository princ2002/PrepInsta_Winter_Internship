```python
# Frog_leap_Problem_Statement

# Step 1: Initiliaze the position List 
position = ['G','G','G','-','B','B','B']
print(position)
```

```python
# Step 2: Initial State of the Game 
print('[  0   1    2    3    4    5    6 ]')
print(position)
```

```python
# Game Loop 
while True:
    # Step 3: Take the input from the user 
    print()
    pos = input("Press 'q' to Quit the game or Enter position of piece:")
    print("Press 'q' to Quit the game \nEnter position of piece:",pos)
    
    # Check if the user wants to quit
    if pos.lower() == 'q': # this fn convert the capital q into small q.
        print('OOOPS! You LoseðŸ¥²ðŸ’” Try Again')
        break # It terminate the loop and Exit the loop if the users quit 
    
    # Step 5: Convert the input into integer 
    pos = int(pos)
    
    # Step 6: Check for valid move 
    if pos <0 or pos >6:
        print('Invalid move')
        continue
    
    # Step 7: Check if the selected position contains a frog.
    if position[pos] == '-':
        print('Invalid move')
        continue
    
    # Step 8: Initialize the the variable empty_index to store the index of the empty leaf 
    pos2 = 0 # position where frog makes jump
    
    # Step 9: Check if the selected frog is 'G'(Green)
    if position[pos] == 'G':
        
        # Condition 1:
        if pos + 1 <= 6 and position[pos+1] =='-':
            pos2 = pos+1
        # Condition 2:
        elif pos + 2 <= 6 and position[pos+2] =='-':
            pos2 = pos + 2 
        # Condition 3:
        else:
            print('Invalid move')
            continue 
    
    # Step 10: Check if he selected frog is 'B'( Brown)
    elif position[pos] == 'B':
        # Condition 1:
        if pos - 1 >= 0 and position[pos-1]=='-':
            pos2 = pos - 1
        # Condition 2:
        elif pos -2 >= 0 and position[pos-2] == '-':
            pos2 = pos - 2
        # Condition 3:
        else:
            print('Invalid Move')
            continue
        
    # Step 11: Swap the elements at the selected position and empty Index
    
    position[pos],position[pos2] = position[pos2],position[pos]
    
    # Step 12: Display the updated Game state
    print('[  0   1    2    3    4    5    6 ]')     
    print(position) 
    # Step 13: Check the winning condition
    if position == ['B','B','B','-','G','G','G']:
        print('HurreyðŸ¥³ðŸŽ‰ You Win!')
        break
    
    # Lossing conditions
    if position == ['-','G','G','G','B','B','B']:
        print('OOOPS! You LoseðŸ¥²ðŸ’” Try Again')
        break
    if position == ['G','G','G','B','B','B','-']:
        print('OOOPS! You LoseðŸ¥²ðŸ’” Try Again')
        break
               
        
        
```
